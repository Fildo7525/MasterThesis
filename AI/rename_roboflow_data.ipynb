```python
import os
import re

FOLDERS_images = [
    "/home/samuel/Downloads/download/full_dataset_new_version/dataset_new_version/combined_dataset_normalized/sorted/images/test",
    "/home/samuel/Downloads/download/full_dataset_new_version/dataset_new_version/combined_dataset_normalized/sorted/images/train",
    "/home/samuel/Downloads/download/full_dataset_new_version/dataset_new_version/combined_dataset_normalized/sorted/images/val",
]


FOLDERS_labels = [
    "/home/samuel/Downloads/download/full_dataset_new_version/dataset_new_version/combined_dataset_normalized/sorted/labels/test",
    "/home/samuel/Downloads/download/full_dataset_new_version/dataset_new_version/combined_dataset_normalized/sorted/labels/train",
    "/home/samuel/Downloads/download/full_dataset_new_version/dataset_new_version/combined_dataset_normalized/sorted/labels/val",
]
```

```python

pattern = re.compile(
    r"(Bjornkjaervej_TestFlight_2)_(bigger|mid|small)_yolov12_(tile_\d+_\d+).*\.jpg"
)

for FOLDER in FOLDERS_images:
    for filename in os.listdir(FOLDER):
        match = pattern.match(filename)
        if not match:
            continue

        base, size, tile = match.groups()
        new_name = f"{base}_{size}_{tile}.jpg"

        old_path = os.path.join(FOLDER, filename)
        new_path = os.path.join(FOLDER, new_name)

        os.rename(old_path, new_path)
        print(f"{filename} -> {new_name}")
```
