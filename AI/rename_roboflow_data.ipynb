```python
import os
import re

FOLDERS_images = [
   "/home/samuel/Downloads/new_dataset/Bjornkjaervej_TestFlight_2_bigger_obb/train/images",
    "/home/samuel/Downloads/new_dataset/Bjornkjaervej_TestFlight_2_mid_obb/train/images",
    "/home/samuel/Downloads/new_dataset/Bjornkjaervej_TestFlight_2_small_obb/train/images",
]


FOLDERS_labels = [
    "/home/samuel/Downloads/new_dataset/Bjornkjaervej_TestFlight_2_bigger_obb/train/labels",
    "/home/samuel/Downloads/new_dataset/Bjornkjaervej_TestFlight_2_mid_obb/train/labels",
    "/home/samuel/Downloads/new_dataset/Bjornkjaervej_TestFlight_2_small_obb/train/labels",
]
```

```python

pattern = re.compile(
    r"(Bjornkjaervej_TestFlight_2)_(bigger|mid|small)_obb_(tile_\d+_\d+).*\.jpg"
)

for FOLDER in FOLDERS_images:
    for filename in os.listdir(FOLDER):
        match = pattern.match(filename)
        if not match:
            continue

        base, size, tile = match.groups()
        new_name = f"{base}_{size}_{tile}.jpg"

        old_path = os.path.join(FOLDER, filename)
        new_path = os.path.join(FOLDER, new_name)

        os.rename(old_path, new_path)
        print(f"{filename} -> {new_name}")
```

```python

pattern = re.compile(
    r"(Bjornkjaervej_TestFlight_2)_(bigger|mid|small)_obb_(tile_\d+_\d+).*\.txt"
)

for FOLDER in FOLDERS_labels:
    for filename in os.listdir(FOLDER):
        match = pattern.match(filename)
        if not match:
            continue

        base, size, tile = match.groups()
        new_name = f"{base}_{size}_{tile}.txt"

        old_path = os.path.join(FOLDER, filename)
        new_path = os.path.join(FOLDER, new_name)

        os.rename(old_path, new_path)
        print(f"{filename} -> {new_name}")
```
